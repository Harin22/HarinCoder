1. Problem Statement: Measures of Central Tendency-I Write a Python program to declare a 1-D array. With the help of the statistics library, calculate the Mean, Median, and Mode

```python
import numpy as np
import pandas as pd
import statistics

i = [1,2,3,4,5,6,2,3,4,5,5,5,5]

mean = statistics.mean(i)
print("Mean:", mean)

median = statistics.median(i)
print("Median:", median)

mode = statistics.mode(i)
print("Mode:", mode)
```

2. Problem Statement: Measures of Central Tendency-IIWrite a Python program to declare a DataFrame as shown in the dataset section below and calculate the Mean, Median, and Mode of the entire DataFrame.

```python
import numpy as np
import pandas as pd
import statistics

i = pd.read_csv("Books.csv")

display(i)

df = pd.DataFrame(i)
display(df)

mean = df.mean()
print("Mean:")
print(mean)

median = df.median()
print("\nMedian:")
print(median)

mode = df.mode()
print("\nMode:")
print(mode)
```
