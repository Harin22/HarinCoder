1. Write a Python program with two DataFrames, with columns “name” and “age”. Now,  combine  both  the  DataFrames  vertically  using  the concat()function, without the duplicate rows, and print the final DataFrame

```python
import numpy as np
import pandas as pd


df1 = pd.DataFrame({
    "name": ["shubham","adin","rahul"],
    "age": [23,28,30,]})

df2 = pd.DataFrame({
    "name": ["amrit","shubham","ali"],
    "age": [20,23,28,]})

print(df1)
print(df2)

combain_df = pd.concat([df1,df2], axis = 0)
print(combain_df)

combain_df.drop_duplicates().reset_index(drop = True)

```

Problem Statement: Combining DataFrames -IIWrite  a  Python  programto  create  two  DataFrames.  First  DataFrame  with columns “employee”and “department,”Second  DataFrame  with  columns “employee”and “joining_date,” and Perform the following tasks:1.Merge the DataFrames based on the “employee”column2.Display the final DataFrame with employees records who are working in the “Engineering” department

```python
import pandas as pd


df1 = pd.DataFrame({
    "employee": ["Lisa","bob","jake","sue","john","daisy"],
    "joining date": [2004,2008,2012,2022,2021,2020]})


df2 = pd.DataFrame({
    "employee": ["Lisa","bob","jake","sue","john","daisy"],
    "department": ["Accounting","engineering","engineering","HR","engineering","engineering"]})


display(df1)
display(df2)
```

2.Merge the DataFrames based on the “employee”column:

```python
merge = pd.merge(df1,df2, on="employee")
print(merge)
```

3.Display   the   final   DataFrame   with   employees   from   the “Engineering” department:


```python
eng_only = merge[merge["department"] == "engineering"]

print(eng_only)
```
