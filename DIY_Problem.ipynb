Problem   Statement: Data   Pre-Processing   (Categorical Data)You are given a dataset –“hotel_bookings.csv.

”The dataset has a high number of null and  elements that need to be cleansed;  Your job is to create a separate DataFrame with only categorical columns and  perform the  following operations

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("hotel_bookings.csv")
df
```

Find the number of null values in each column of the new DataFr

```python
cat_cols = df.select_dtypes(include=['object'])
cat_df = pd.DataFrame(cat_cols)
cat_df
```

Replace the null values with mode 

```python
null_counts = cat_df.isnull().sum()
print(null_counts)
```

In the "hotel"column, replace the hotel names with "0" and "1" based on the condition that –if,"hotel" = "city_hotel",then "hotel"= "1";  else, "0"

```python
cat_df.loc[:, 'hotel'] = cat_df.loc[:, 'hotel'].apply(lambda x: "1" if x == "City Hotel" else "0")

```

Using the label encoder, assign a unique country code to each co

```python
le = LabelEncoder()
cat_df['country_code'] = le.fit_transform(cat_df['country'])
print(cat_df.head())
```

Using onehot encoder, encode the month column

```python
month_dummies = pd.get_dummies(cat_df['arrival_date_month'])
print(month_dummies.head())
```
