1.   Problem   Statement: Splitting   Data   for   Testing   and TrainingYou  are  given  a  dataset –“housing.csv.”  Load  the  dataset  into  a DataFrame.   

 Considering    the “median_house_value”column    as output/target  and  the  rest  of  the  columns  as  input,  perform  the  following tasks:
 
 1.Find  the  number  of  null  values  in  each  column  and  replace themwith  the mean values

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('housing.csv')
df
```

 2.Split  the  data  into  test  and  train  fragments  using train_test_split()function in 80:20 ratio (80% train, 20% test)

```python
df.fillna(df.mean(), inplace=True)
```

 3.Print the size of test and train data (For both input and output)

```python
train_data, test_data, train_target, test_target = train_test_split(df.drop('median_house_value', axis=1), 
                                                                    df['median_house_value'], 
                                                                    test_size=0.2, random_state=42)
print('Size of train data:', train_data.shape)
print('Size of train target:', train_target.shape)
print('Size of test data:', test_data.shape)
print('Size of test target:', test_target.shape)
```
